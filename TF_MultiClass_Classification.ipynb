# Conver target in to numerical data

```python
from tensorflow.keras.utils import to_categorical
```

```python
y_train_ohe = to_categorical(y_train)
y_test_ohe = to_categorical(y_test)
```

```python
print(x_train.shape)
print(y_train_ohe.shape)
print(x_test.shape)
print(y_test_ohe.shape)
```

### Model(NN) Building

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense #Fully connected layer
from tensorflow.keras.layers import Input
from tensorflow.keras.optimizers import Adam,Adagrad,Adadelta
from tensorflow.keras.regularizers import L1L2

model = Sequential()
model.add(Input(shape=(784,))) #No computation here-This is not mandatory
# model.add(Dense(units=100, activation='relu') #HL
model.add(Dense(units=100, activation='relu',kernel_regularizer=L1L2(l1=0,l2=0.1))) #HL
model.add(Dense(units=10,activation='softmax')) 
```

### Ex: For softmax

```python

import numpy as np
```

```python
print((np.exp(-10)/(np.exp(-10)+np.exp(5)+np.exp(3))))
print((np.exp(5)/(np.exp(-10)+np.exp(5)+np.exp(3))))
print((np.exp(3)/(np.exp(-10)+np.exp(5)+np.exp(3))))
```

```python
model.compile(optimizer=Adam(),loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(
    x=x_train,
    y=y_train_ohe,
    epochs=100,
    validation_data=(x_test, y_test_ohe),
    batch_size=1000
)
```

```python
import matplotlib.pyplot as plt

train_loss = model.history.history['loss']
val_loss = model.history.history['val_loss']

plt.plot(train_loss)
plt.plot(val_loss)
plt.ylim([0,6]) 
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend(['Train','Val'])
plt.title("Loss Curve")
plt.grid()
plt.show()
```

```python
# Extract accuracy
train_acc = model.history.history['accuracy']
val_acc = model.history.history['val_accuracy']

# Plot accuracy
plt.plot(train_acc)
plt.plot(val_acc)
plt.ylim((0.5,1))
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend(['Train', 'Validation'])
plt.title("Accuracy Curve")
plt.grid(True)
plt.show()
```
