```python
x1=[10, 8, 15]
x2=[8, 20, 25]
y=[20, 15, 30]


n=len(x1)# num of observations
alpa=0.01
```

```python
import numpy as np
np.random.seed(123)
random_init=np.random.randn(3)
w1, w2, b = random_init[0], random_init[1], random_init[2]

print(f"Initialized weights-> w1 : {w1} | w2 : {w2} | b : {b} ")
```

```python
#Finding MSE
y_pred= []
total_error=0

for i in range(n):
    y_hat=w1*x1[i]+w2*x2[i]+b
    y_pred.append(y_hat)#Predected values

    E= (y_hat-y[i])**2
    total_error=total_error+E
```

```python
print(y_pred)
print()
print(total_error/n)#MSE
```

```python
def forward_pass(w1,w2,b):
    y_pred= []
    total_error=0

    for i in range(n):
        y_hat=w1*x1[i]+w2*x2[i]+b
        y_pred.append(y_hat)#Predected values

        E= (y_hat-y[i])**2
        total_error=total_error+E
    return y_pred,total_error/n
```

```python
def weight_update(w1,w2,b,y_pred):
    for i in range(n):
        dmse_dw1 = 2 * (y_pred[i]-y[i]) * x1[i]
        dmse_dw2 = 2 * (y_pred[i]-y[i]) * x2[i]
        dmse_db = 2 * (y_pred[i]-y[i]) 

        w1 = w1 - alpa * dmse_dw1
        w2 = w2 - alpa * dmse_dw2
        b = b - alpa * dmse_db
    
    return w1,w2,b
```

```python
#Initilization
print(f"Initialized weights-> w1 : {w1} | w2 : {w2} | b : {b} ")

#Iteration 1

#Step 1:Forward_pass
y_pred,total_error = forward_pass(w1,w2,b)
print(f"Iteration 1 -> Error : : {total_error}")

#Step 2:Weight_update
w1, w2 ,b=weight_update(w1,w2,b,y_pred)
print(f"Updated weights-> w1 : {w1} | w2 : {w2} | b : {b} ")

print()

#Iteration 2

#Step 1:Forward_pass
y_pred,total_error = forward_pass(w1,w2,b)
print(f"Iteration 2 -> Error : : {total_error}")

#Step 2:Weight_update
w1, w2 ,b=weight_update(w1,w2,b,y_pred)
print(f"Updated weights-> w1 : {w1} | w2 : {w2} | b : {b} ")

print()

#Iteration 3

#Step 1:Forward_pass
y_pred,total_error = forward_pass(w1,w2,b)
print(f"Iteration 3 -> Error : : {total_error}")

#Step 2:Weight_update
w1, w2 ,b=weight_update(w1,w2,b,y_pred)
print(f"Updated weights-> w1 : {w1} | w2 : {w2} | b : {b} ")
```
